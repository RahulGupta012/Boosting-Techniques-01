# Q1. What is boosting in machine learning?

Boosting in machine learning is a ensable technique which aims to increase the performance of the model. In the boosting technique there are weak learners (models) which squencially arranged in order to accelerate the performance of the model. Each weak learner transfer the data to another waek learner, who is more better than previous one, to capture the true information as per the capibility of the weak learner. And at the end of the process, when the data has going in the authority of strong learner, then it will give us the final outcome.There are commnley three type of boosting in ensamble techniques :

1. Adaboost Boosting
2. Gradient Boosting
3. Xgboost Boosting

There are some steps by which boosting takes place ;

**Step 01 :** ***Assign equal weights*** to all training instances.

**Step 02 :** Fit a weak model on the ***training data***with the weights assigned to each instance.

**Step 03 :** ***Evaluate the performance*** of the weak model and compute the errors.

**Step 04 :**  Increase the weights of misclassified instances, so they become more important in the next iteration. Decrease the weights of correctly classified instances. Which means we have ***update the wights***.

**Step 05 :**  ***Combine the weak models*** into a strong model by giving more weight to models that performed well in terms of classification accuracy.

  **n_estimators:** This parameter determines the number of weak learners to be trained in the ensemble. Increasing the number of estimators can improve performance, but it also increases the computational cost.
  
  **max_depth:** The maximum depth of each weak learner . Limiting the depth helps prevent overfitting, but setting it too low may lead to underfitting.
  
  **min_samples_split:** The minimum number of samples required to split an internal node in a weak learner. This parameter helps control the tree's growth and can impact the balance between bias and variance.
  
  **min_samples_leaf:** The minimum number of samples required to be in a leaf node. It influences the granularity of the tree and can also impact overfitting.
  
  **Random State:** A seed or random state to ensure reproducibility. Setting a random state ensures that the randomization during training is consistent across different runs.
